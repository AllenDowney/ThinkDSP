#Synthesizing the birdcall of the tree pipit through visual wave, spectrum, and spectrogram analysis

   Birdcalls are a touchstone sound of nature. However, they are also comprised of the same signal components that make up every other audible sound and as a consequence should be synthesizable through the combination of various waveforms. In this notebook I have explored the process of replicating the sound of the tree pipit through visually analyzing the wave form, spectrum, and spectrogram of the birdcall. I method I undertook involved picking key features of the original signal visible in the different representations and recreating them through the formation of a frequency profile. After concatenating chirps into a full call, modifying the amplitude and duration of each chirp created a decent reproduction of the original call.
   
   The final output, while a decent representation of the original call, is missing two minor features that contribute a significant amount to the realism of the bird call. Those features are the heavily variable amplitudes present in each chirp, and the slight trill present on the tail end of each chirp. Despite these ommissions this exploration still demonstrates the ability to recreate a natural sounds through the synthesis of many mathematical signals.

##Constructing the frequency curve of a single chirp

```python
ts = np.zeros(11025) # An arbitrary length to create samples on
x = np.linspace(0,2*math.pi,len(ts))
i = np.sin(x/2) # A half period of a sine wave over the length of sampling
x2 = np.linspace(0,math.pi,len(ts)/2)
i2 = np.zeros(len(ts))
i2[0:len(x2)] = np.sin(x2) # A half period of a sine wave over the first half of the samples
i3 = np.sin(50*x)/25 # A small magnitude sine wave with high frequency
i5 = np.random.rand(len(ts))/20 # Some random values with small amplitude
i = i + i2 # Adding the two (low frequency) sine waves
i = i/2.3 + 0.6 # Squishing the curve and shifting the baseline up
wav = thinkdsp.Wave(ys = i, framerate=len(ts)/10)
spec = wav.make_spectrum()
spec.low_pass(.6) # Wave -> Spectrum so that I can low pass to smooth the signal out
wav = spec.make_wave() # convert back to wave
if not (len(wav.ys) & 0) and len(ts) & 1:
    wav.ys = np.append(wav.ys, wav.ys[-1]) # Converting to spectrum and back will lose one sample if the sample length is odd
wav.ys = wav.ys + i3 + i5 # Add the high frequency sine and random values
wav.plot()
print(wav.ys[0]) # Ensure the wave is close to periodic to prevent frequency jumps (causing clicks)
print(wav.ys[-1])
```

The above curve is a rough approximation of the frequency shape within one chirp of the birdcall spectrogram. The key components are the sinusoid-ish shape with a low tail, and both consistent and random high frequency components to add minor variety to the call.

## birdChirp class based on the ThinkDSP Chirp class

```python
class birdChirp(thinkdsp.Chirp):
    def evaluate(self, ts):
        # The frequency curve from the above cell
        x = np.linspace(0,2*math.pi,len(ts))
        i = np.sin(x/2)
        x2 = np.linspace(0,math.pi,len(ts)/2)
        i2 = np.zeros(len(ts))
        i2[0:len(x2)] = np.sin(x2)
        i3 = np.sin(50*x)/25
        i5 = np.random.rand(len(ts))/20
        i = i + i2
        i = i/2.3 + 0.6
        wav = thinkdsp.Wave(ys = i, framerate=len(ts)/10)
        spec = wav.make_spectrum()
        spec.low_pass(.6)
        wav = spec.make_wave()
        if not (len(wav.ys) & 0) and len(ts) & 1:
            wav.ys = np.append(wav.ys, wav.ys[-1])
        wav.ys = wav.ys + i3 + i5
        # Set the curve as the frequency pattern of the evaluated wave
        start, end = self.start, self.end
        freqs = np.linspace(start,end, len(ts)-1)*wav.ys[:-1]
        return self._evaluate(ts, freqs)
```

##Generate an 8 chirp birdcall

```python
wave = False
fr = 44100
chirpsPerCall = 8

for i in range(chirpsPerCall):
    # Generate chirps with decreasing start and end frequencies
    sig = birdChirp(start=5200-(i*80), end=4300-((i-1)*80))
    # Calculating an increasing length of the chirps over time
    length = 0.1+(0.008*i**1.5)
    # Create a wave
    waveSeg = sig.make_wave(duration=length, framerate=fr)
    ampMult = np.zeros(len(waveSeg.ys))
    # Create a "triangle" to fade each chirp in and out
    ampMult1 = np.linspace(0,1,math.floor(len(waveSeg.ys)/2))
    ampMult2 = np.linspace(1,0,len(waveSeg.ys)-math.floor(len(waveSeg.ys)/2))
    ampMult = np.concatenate((ampMult1, ampMult2))
    waveSeg.ys = waveSeg.ys*ampMult # Add fades to chirps
    waveSeg.ys = waveSeg.ys*(i/chirpsPerCall+.2) # Increase amplitude of each consecutive chirp
    # Add a slight increasing delay between each chirp
    waveSeg.ys = np.concatenate((waveSeg.ys, np.zeros(math.floor((length/4)*44100))))
    if wave:
        wave = wave | waveSeg # Concat chirps together into single wave
    else:
        wave = waveSeg
```

## Visualizing the results

```python
wave.plot()
```

The waveform roughly approximates the waveform of the original birdcall. Each chirp is still very similar and there is not much of the amplitude variation that exists in the original. 

```python
spec = wave.make_spectrum()
spec.low_pass(7400)
spec.high_pass(2200)
spec.plot()
wave = spec.make_wave()
```

The spectrum is more related to the original spectrum that the waveform. The bulk of the frequencies are grouped around the same frequencies and the two peaks are similar in frequency at about 4800 Hz and 5100 Hz. The frequencies are a bit less spread out than the original and the 4800 Hz peak is wider than the very narrow peak in the original.

```python
gram = wave.make_spectrogram(seg_length=1024)
gram.plot()
```

The spectrogram is very similar to the original. The main shapes and features are recreated fairly accurately. The main difference is a slight increase in frequency variation across each chirp and the slight bump on the tail of the chirp. These differences do not make too substantial of a difference in audible differences. 

```python
wave.apodize()
wave.make_audio()
```

The final audio has a fairly close resemblance to the original. Some of the most noticeable differences from the original are the more linear increases in duration and the lack of the occassional trill. While without these features the song is still recognizable as a birdcall, the addition of these last features would add an additional element of realism that would help differentiate the sound from an algorithmically generated signal.
